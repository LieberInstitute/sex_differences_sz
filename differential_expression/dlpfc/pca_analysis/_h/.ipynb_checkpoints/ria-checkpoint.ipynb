# PCA Analysis

```python
#load required packages
import os, errno
import numpy as np
import pandas as pd
from plotnine import *
from collections import Counter
from scipy.stats import spearmanr
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from warnings import filterwarnings
from matplotlib.cbook import mplDeprecation

filterwarnings("ignore",category=mplDeprecation)
```

## Load phenotype data

```python
phenotype_file = '../../_m/caudate_phenotypes.csv'
pheno = pd.read_csv(phenotype_file, index_col=0)
pheno.head()
```

## Functions

```python
def extract_deg(filename, thres=0.05):
    ''' Take DE genes obtained from limma-voom pipeline, subset index, 
        gencodeID, logFC, t-statistic, p + FDR p vals.
    '''
    de_df = pd.read_csv(filename, sep='\t') # Load DE file
    deg = de_df[(de_df['adj.P.Val'] < thres)] # Subset adjusted p value by threshold
    deg = deg[['Geneid','logFC', 't', 'P.Value', 'adj.P.Val']] # Select rows of interest
    return deg


def load_data(inputfile, filename1, filename2):
    '''Load residualized expression file from limma-voom pipeline for 
       all DE genes, load adjusted DE gene table
    '''
    res = pd.read_csv(inputfile, sep='\t', index_col=0).transpose()
    de_df = extract_deg(filename1)
    noX_df = extract_deg(filename2)
    return res, de_df, noX_df


def load_res(res, df):
    '''Generates a residualized expression dataframe for significant 
       DE genes only (FDR < 0.05)
    '''
    resdf = res[df.Geneid]
    return resdf


def calculate_corr(xx, yy, lab):
    '''This calculates Spearman correlation:
         - used to calculate monotonicity of the relation of 2 datasets 
         - the datasets are principal components 1 and 2 (PC1, PC2)
         - calculated on a scale of -1, 1 (0 being no correlation)
         - figure out monotonicity between PC1, PC2
         - figure out monotonicity between PC2, PC1
        Inputs: PC1, PC2, label
        Outputs: 
          1. rho (correlation coefficient)
          2. p-value, two-sided test 
            - whose null hypothesis is that two sets of data are uncorrelated
        Notes: p-value has same dimension as rho
    '''
    rho, pvalue = spearmanr(xx, yy)
    print('%s - Rho: %.4f, PValue: %.2e' % (lab, rho, pvalue)) 

    
def cal_pca(res, df, num, pheno):
    '''
    num: refers to how many genes we want to include
    if 0, look at all genes, if not 0, look at only those specified genes
    new_pheno: this is the correct size of samples using the the first two columns of residualized expression
      - the residualized expression data frame, has the correct samples (dropping qSV samples included in pheno)
      - output new_pheno shape row numbers should be the same as res_df row numbers
    '''
    if num == 0:
        expr_res = load_res(res, df)
    else:
        expr_res = load_res(res, df.head(num))
    
    new_pheno = pheno.merge(expr_res.iloc[:, 0:1], right_index=True, # Generate pheno data frame with correct samples
                            left_index=True).drop(expr_res.iloc[:, 0:1].columns, axis=1)

    x = StandardScaler().fit_transform(expr_res)
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data=principalComponents, 
                               columns=['PC1', 'PC2'], 
                               index=expr_res.index)
    finalDf = pd.concat([principalDf, new_pheno], axis = 1)
    return finalDf


def plot_corr_save(res, df, num, label, pheno=pheno):
    pca_df = cal_pca(res, df, num, pheno)
    #pp = ggplot(pca_df, aes(x='PC1', y='PC2', color='sex')) + geom_point() + theme_light()
    pp = ggplot(pca_df, aes(x='PC1', y='PC2', color='sex')) + geom_point() + theme_light() + theme(
                         axis_text_x=element_blank(), axis_text_y=element_text(size=18), 
                         axis_title_y=element_text(size=21), axis_title_x=element_text(size=21), 
                         legend_text=element_text(size=16), legend_title=element_blank())
    print(pp)
    if num == 0:
        pp.save('%s_pca_all.pdf' % (label), width = 5, height = 8)
        pp.save('%s_pca_all.svg' % (label), width = 5, height = 8)
        pp.save('%s_pca_all.png' % (label), width = 5, height = 8)
    else:
        pp.save('%s_pca_top_%d.pdf' % (label, num), width = 5, height = 8)
        pp.save('%s_pca_top_%d.svg' % (label, num), width = 5, height = 8)
        pp.save('%s_pca_top_%d.png' % (label, num), width = 5, height = 8)
    print(calculate_corr(pca_df.PC1, pca_df.sex, 'PC1'))
    print(calculate_corr(pca_df.PC2, pca_df.sex, 'PC2'))
    
```

```python
def plot_corr(res, df, num, label, pheno=pheno):
    pca_df = cal_pca(res, df, num, pheno)
    pp = ggplot(pca_df, aes(x='PC1', y='PC2', color='sex')) + geom_point() + theme_light() + theme(
                         axis_text_x=element_blank(), axis_text_y=element_text(size=18), 
                         axis_title_y=element_text(size=21, face='bold'), axis_title_x=element_blank(), 
                         legend_text=element_text(size=16), legend_title=element_blank())
    print(pp)
    print(calculate_corr(pca_df.PC1, pca_df.sex, 'PC1'))
    print(calculate_corr(pca_df.PC2, pca_df.sex, 'PC2'))
    
```
