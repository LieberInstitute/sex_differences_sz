# Venn diagram and summary

```python
import re
import numpy as np
import pandas as pd
from venn import venn
from matplotlib import pyplot as plt
```

## Prepare data

### Genes

```python
genes = pd.read_csv('../../_m/genes/diffExpr_szVctl_full.txt', sep='\t')
genes = genes[(genes['adj.P.Val'] < 0.05)].sort_values('adj.P.Val')
genes['Feature'] = genes.index
genes = genes[['Feature', 'Symbol', 'ensemblID', 'logFC', 'adj.P.Val']]
genes['Type'] = 'gene'
genes.head()
```

### Transcripts

```python
trans = pd.read_csv('../../_m/transcripts/diffExpr_szVctl_full.txt', sep='\t')
trans = trans[(trans['adj.P.Val'] < 0.05)].sort_values('adj.P.Val')
trans['Feature'] = trans.index
trans['ensemblID'] = trans.gene_id.str.replace('\\.\d+', '')
trans = trans[['Feature', 'gene_name', 'ensemblID', 'logFC', 'adj.P.Val']].rename(columns={'gene_name': 'Symbol'})
trans['Type'] = 'transcript'
trans.head()
```

### Exons

```python
exons = pd.read_csv('../../_m/exons/diffExpr_szVctl_full.txt', sep='\t')
exons = exons[(exons['adj.P.Val'] < 0.05)].sort_values('adj.P.Val')
exons['Feature'] = exons.index
exons = exons[['Feature', 'Symbol', 'ensemblID', 'logFC', 'adj.P.Val']]
exons['Type'] = 'exon'
exons.head()
```

### Junctions

```python
juncs = pd.read_csv('../../_m/junctions/diffExpr_szVctl_full.txt', sep='\t')
juncs = juncs[(juncs['adj.P.Val'] < 0.05)].sort_values('adj.P.Val')
juncs['Feature'] = juncs.index
juncs = juncs[['Feature', 'Symbol', 'ensemblID', 'logFC', 'adj.P.Val']]
juncs['Type'] = 'junction'
juncs.head()
```

## Differential Expression Summary

### Feature summary
#### DE

```python
gg = len(set(genes['Feature']))
tt = len(set(trans['Feature']))
ee = len(set(exons['Feature']))
jj = len(set(juncs['Feature']))

print("\nGene:\t\t%d\nTranscript:\t%d\nExon:\t\t%d\nJunction:\t%d" % (gg, tt, ee, jj)) 
```

#### DE (EnsemblID)

```python
gg = len(set(genes['ensemblID']))
tt = len(set(trans['ensemblID']))
ee = len(set(exons['ensemblID']))
jj = len(set(juncs['ensemblID']))

print("\nGene:\t\t%d\nTranscript:\t%d\nExon:\t\t%d\nJunction:\t%d" % (gg, tt, ee, jj)) 
```

#### DE (Gene Symbol)

```python
gg = len(set(genes['Symbol']))
tt = len(set(trans['Symbol']))
ee = len(set(exons['Symbol']))
jj = len(set(juncs['Symbol']))

print("\nGene:\t\t%d\nTranscript:\t%d\nExon:\t\t%d\nJunction:\t%d" % (gg, tt, ee, jj)) 
```

### Feature effect size summary

```python
feature_list = ['Genes', 'Transcript', 'Exons', 'Junctions']
feature_df = [genes, trans, exons, juncs]
for ii in range(4):
    ff = feature_df[ii]
    half = len(set(ff[(np.abs(ff['logFC']) >= 0.5)].Feature))
    one = len(set(ff[(np.abs(ff['logFC']) >= 1)].Feature))
    print("\nThere are %d unique %s with abs(log2FC) >= 0.5" % (half, feature_list[ii]))
    print("There are %d unique %s with abs(log2FC) >= 1" % (one, feature_list[ii]))
```

```python
feature_list = ['Genes', 'Transcripts', 'Exons', 'Junctions']
feature_df = [genes, trans, exons, juncs]
for ii in range(4):
    ff = feature_df[ii]
    half = len(set(ff[(np.abs(ff['logFC']) >= 0.5)].ensemblID))
    one = len(set(ff[(np.abs(ff['logFC']) >= 1)].ensemblID))
    print("\nThere are %d unique %s with abs(log2FC) >= 0.5" % (half, feature_list[ii]))
    print("There are %d unique %s with abs(log2FC) >= 1" % (one, feature_list[ii]))
```

## Plot venn diagram

```python
features = {
    'Genes': set(genes['ensemblID']),
    'Transcripts': set(trans['ensemblID']),
    'Exons': set(exons['ensemblID']),
    'Junctions': set(juncs['ensemblID']),
}
```

```python
venn(features, fmt="{percentage:0.1f}%", fontsize=20, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage.png')
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage.pdf')
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage.svg')
plt.show()
```

```python
venn(features, fmt="{size}\n{percentage:0.1f}%", fontsize=16, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage_size.png')
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage_size.pdf')
plt.savefig('DE_genes_exons_juncs_venn_diagram_percentage_size.svg')
plt.show()
```

### Upregulated

```python
features_up = {
    'Genes': set(genes[(genes['logFC'] > 0)].loc[:, 'ensemblID']),
    'Transcripts': set(trans[(trans['logFC'] > 0)].loc[:, 'ensemblID']),
    'Exons': set(exons[(exons['logFC'] > 0)].loc[:, 'ensemblID']),
    'Junctions': set(juncs[(juncs['logFC'] > 0)].loc[:, 'ensemblID']),
}
```

```python
venn(features_up, fmt="{percentage:0.1f}%", fontsize=20, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage.png')
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage.pdf')
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage.svg')
plt.show()
```

```python
venn(features_up, fmt="{size}\n{percentage:0.1f}%", fontsize=16, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage_size.png')
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage_size.pdf')
plt.savefig('DE_upregulated_genes_exons_juncs_venn_diagram_percentage_size.svg')
plt.show()
```

### Downregulated

```python
features_down = {
    'Genes': set(genes[(genes['logFC'] < 0)].loc[:, 'ensemblID']),
    'Transcripts': set(trans[(trans['logFC'] < 0)].loc[:, 'ensemblID']),
    'Exons': set(exons[(exons['logFC'] < 0)].loc[:, 'ensemblID']),
    'Junctions': set(juncs[(juncs['logFC'] < 0)].loc[:, 'ensemblID']),
}
```

```python
venn(features_down, fmt="{percentage:0.1f}%", fontsize=20, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage.png')
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage.pdf')
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage.svg')
plt.show()
```

```python
venn(features_down, fmt="{size}\n{percentage:0.1f}%", fontsize=16, legend_loc="best", figsize=(12, 12))
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage_size.png')
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage_size.pdf')
plt.savefig('DE_downregulated_genes_exons_juncs_venn_diagram_percentage_size.svg')
plt.show()
```

## Directional analysis

### All significant DEG

```python
up = len(features_up['Genes']) / len(features['Genes']) * 100
down = len(features_down['Genes']) / len(features['Genes']) * 100
print("The percentage of up-regulated: %0.2f%%" % up)
print("The percentage of down-regulated: %0.2f%%" % down)
```

### Top 100 DEG

```python
top_genes = genes.sort_values('adj.P.Val').head(100)
feat_all = set(top_genes.ensemblID)
feat_up = set(top_genes[(top_genes['logFC'] > 0)].loc[:, 'ensemblID'])
feat_down = set(top_genes[(top_genes['logFC'] < 0)].loc[:, 'ensemblID'])
up = len(feat_up) /  len(feat_all) * 100
down = len(feat_down) /  len(feat_all) * 100
print("The percentage of up-regulated: %0.2f%%" % up)
print("The percentage of down-regulated: %0.2f%%" % down)
```

## Percent overlap by limiting feature

```python
def limiting_features(f1, f2):
    xx = len(features[f1] & features[f2]) / len(features[f2]) * 100
    print("Comparing %s with %s: %0.2f%%" % (f1, f2, xx))
    print("Features in common: %d" % len(features[f1] & features[f2]))
```

### Gene comparison

```python
limiting_features('Genes', 'Transcripts')
limiting_features('Genes', 'Junctions')
limiting_features('Exons', 'Genes')
```

### Transcript comparison

```python
limiting_features('Exons', 'Transcripts')
limiting_features('Junctions', 'Transcripts')
```

### Exon comparison

```python
limiting_features('Exons', 'Junctions')
```
