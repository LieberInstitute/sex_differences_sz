# Summary of results

```r
library(data.table)
```

```r
de_fdr.file <- '/ceph/users/jbenja13/ria_tmp/projects_version2/sex_differences_sz/analysis/sex_differences//_m/diffExpr_gender_FDR05.txt'
de_fdr <- read.csv(de_fdr.file, sep = '\t')
head(de_fdr)
```

```r
noX <- de_fdr[! (de_fdr$Chr %like% "chrX"),]
dim(noX)
```

```r
aa <- noX 
aa$Chr <- as.vector(noX$Chr)
```

```r
noY <- de_fdr[! (de_fdr$Chr %like%"chrY"),]
dim(noY)
```

```r
bb <- noY
bb$Chr <- as.vector(noY$Chr)
```

```r
noXY <- data.frame(intersect(aa$Chr, bb$Chr))
dim(noXY)
```

```r
noXY_df <- merge(de_fdr, noXY, by.x = "Chr", by.y = 'intersect.aa.Chr..bb.Chr.')
head(noXY_df)
dim(noXY_df)
```

```r
write.table(noXY_df, file = 'diffExpr_FDR05_noXY.txt', sep='\t', row.names=FALSE, quote=FALSE)
```
